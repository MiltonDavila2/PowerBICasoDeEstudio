```python
import pandas as pd
```

```python
file_path= 'Appol Datos.xlsx'

datos = pd.read_excel(file_path,sheet_name="Tabla Datos")
productos = pd.read_excel(file_path,sheet_name="Tabla Productos")
```

```python
datos
```

```python
productos
```

```python
datos[['Qs', 'Año']] = datos['Periodo'].str.extract(r'(Q\d)\s+(\d{4})')
```

```python
datos
```

```python
region_split = datos['Region'].str.extract(r'-\s*(.*?)\/(.*)')

# Crear las nuevas columnas
datos['Continente'] = region_split[0].str.strip().str.capitalize()  # Ej: "África"
datos['Pais'] = region_split[1].str.strip().str.capitalize()        # Ej: "Angola"
```

```python
datos
```

```python
datos = datos.drop(columns=['Region'])
```

```python
datos = datos.merge(productos, on="Tipo Producto", how="left")
```

```python
output_file_path = 'Tabla_Datos_Modificada.xlsx'
datos.to_excel(output_file_path, index=False)

print(f'✅ Archivo guardado en: {output_file_path}')
```
